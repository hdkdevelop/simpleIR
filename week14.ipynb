```python
import matplotlib.pyplot as plt
import csv
import numpy as np

f = open('./202012_202012_population.csv', encoding='euc-kr')
data = csv.reader(f)
next(data)
next(data)

# 광진구
criteria = 5
datas = list(data)
diffList = []
for index, row in enumerate(datas): 
    if index != criteria:
        currentDiff = 0
        for j in range(3, len(row)):
            currentDiff += np.square(int(row[j].replace(',','')) - int(datas[criteria][j].replace(',','')))
        diffList.append([currentDiff, index])
        
diffList.sort()

for index, row in enumerate(datas): 
    if index == criteria:
        row = row[3:]
        result = [int(i.replace(',','')) for i in row]
        plt.plot(result)
    if index == diffList[0][1]:
        label = row[0]
        row = row[3:]http://127.0.0.1:8888/notebooks/Desktop/konkuk/%E1%84%8B%E1%85%A9%E1%84%91%E1%85%B3%E1%86%AB%E1%84%89%E1%85%A9%E1%84%89%E1%85%B3SW/simpleIR/week14.ipynb#
        result = [int(i.replace(',','')) for i in row]
        plt.plot(result, label=label)
    if index == diffList[-1][1]:
        label = row[0]
        row = row[3:]
        result = [int(i.replace(',','')) for i in row]
        plt.plot(result, label=label)
plt.style.use('ggplot')
plt.rc('font', family='AppleGothic')
plt.legend(loc='upper right')
plt.show()
```
