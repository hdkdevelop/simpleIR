```python
import matplotlib.pyplot as plt
import csv

f = open('./seoul_from2011.csv', encoding='euc-kr')
data = csv.reader(f)
next(data)
maxTemp = []
minTemp = []
for row in data :
    if row[-1] != '':
        if row[0].split('.')[1] == '8' and row[0].split('.')[2] == '4' :
            maxTemp.append(float(row[-1]))
            minTemp.append(float(row[-2]))
plt.plot(maxTemp, color='hotpink')
plt.plot(minTemp, color='skyblue')
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('./seoul_from2011.csv', encoding='euc-kr')
data = csv.reader(f)
next(data)
janTemp = []
augTemp = []
for row in data :
    if row[-1] != '':
        month = row[0].split('.')[1]
        if month == '1':
            janTemp.append(float(row[-1]))
        if month == '8':
            augTemp.append(float(row[-1]))
plt.hist(janTemp, color='b', bins=100, label="Jan")
plt.hist(augTemp, color='r', bins=100, label="Aug")
plt.legend(loc='upper right')
plt.show()
```
